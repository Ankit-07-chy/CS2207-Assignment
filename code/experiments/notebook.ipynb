```python
# Importing 
import os 
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the data
from pathlib import Path

# Find the repository root by walking upward until a known file exists.
root = Path.cwd().resolve()
while not (root / 'pyproject.toml').exists() and root != root.parent:
    root = root.parent

file_path = root / 'data' / 'credit_card_usage.csv'
data = pd.read_csv(file_path)
```

```python
# Showing the first few rows of the dataset
display(data.head())
print("#"*100)
print(data.info())
```

```python
# Columns Uisng
data.columns
```

```python
data = data.drop('CUST_ID', axis=1)
```

```python
data.isnull().sum()
```

```python
# Taking Care of Missing Values
data['MINIMUM_PAYMENTS'] = data['MINIMUM_PAYMENTS'].fillna(
    data['MINIMUM_PAYMENTS'].median()
)

data.isnull().sum()
```

```python
data.duplicated().sum()
```

```python
# CREDIT LIMIT remove none
data = data.dropna(subset=['CREDIT_LIMIT'])
```

```python
# For Transforming the data, we will use the log transformation to reduce the skewness of the data.
skewed_cols = ['BALANCE', 'PURCHASES', 'ONEOFF_PURCHASES',
               'INSTALLMENTS_PURCHASES', 'CASH_ADVANCE',
               'PAYMENTS', 'MINIMUM_PAYMENTS']

for col in skewed_cols:
    data[col] = np.log1p(data[col])
```

```python
# Feature Scaling 
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)
```

```python
scaled_data = pd.DataFrame(scaled_data, columns=data.columns)
```

```python
scaled_data
```

```python
# Test different MinPts values
from sklearn.neighbors import NearestNeighbors
for min_pts in [3, 5, 10, 15]:
    neighbors = NearestNeighbors(n_neighbors=min_pts)
    neighbors.fit(scaled_data)
    distances, _ = neighbors.kneighbors(scaled_data)
    distances = np.sort(distances[:, min_pts-1])
    plt.plot(distances, label=f'MinPts={min_pts}')

plt.legend()
plt.title("K-Distance Plots for Different MinPts")
plt.show()
```

```python
# k-distance plot to determine the optimal epsilon value for DBSCAN
from sklearn.neighbors import NearestNeighbors
import matplotlib.pyplot as plt
import numpy as np

neighbors = NearestNeighbors(n_neighbors=5)
neighbors_fit = neighbors.fit(scaled_data)
distances, indices = neighbors_fit.kneighbors(scaled_data)

distances = np.sort(distances[:, 3])  # 4 = MinPts-1
plt.plot(distances)
plt.title("K-Distance Plot")
plt.show()
```

```python
# Apply DBSCAN
from sklearn.cluster import DBSCAN

dbscan = DBSCAN(eps=1.7, min_samples=5)  # adjust eps from plot
labels = dbscan.fit_predict(scaled_data)

data['Cluster'] = labels
```
